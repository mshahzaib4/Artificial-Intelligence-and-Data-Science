> **Import library**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

> **Import the data**

```python
matches=pd.read_csv("matches.csv")
matches.head(2)
```

### `1.` **Create COUNT Plot**

> **Create a basic COUNTPLOT to get number of transaction for each of the product line and change the figure size**

```python
plt.figure(figsize=(12,8))
sns.countplot(x="winner",data=matches)
```

> **Add hue to get the count on two categories i.e. winner and match_type**


# **`2.` Create BARPLOT**


```python
plt.figure(figsize=(12,8))
sns.countplot(x="winner",hue="match_type",data=matches)
```

----
----

```python
matches["match_type"].value_counts()
```

## **Create a basic bar plot**


```python
plt.figure(figsize=(12,8))

sns.barplot(data=matches,x="winner",y="winner_runs")
```

### **Add HUE and Hue_order to the barplot**

```python
plt.figure(figsize=(12, 8))

sns.barplot(data=matches, x="winner", y="winner_runs", hue="match_type", hue_order=["Final", "Semi Final","Group"])
plt.show()
```

----
----




### **`3.` Create a basic `BOX PLOT`**


```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=matches, x="winner", hue="match_type", width=0.7)
```

----
----



### **`3.` Create a basic `VIOLIN PLOT`**


```python
plt.figure(figsize=(12, 8))
sns.violinplot(data=matches, x="winner",y="winner_runs",hue="season")
```


### **`4.` Create a basic `STRIP PLOT`**


```python
plt.figure(figsize=(12, 8))
sns.stripplot(data=matches,y="team1",x="team2",hue="match_type",size=10)
```

### **`5.` Create a basic `SWARM PLOT`**


```python
plt.figure(figsize=(12, 8))
sns.swarmplot(data=matches,y="team1",x="team2",hue="match_type",size=10)
```

### **`6.` Create a basic `FACTOR PLOT`**


```python

plt.figure(figsize=(12, 15))

sns.catplot(data=matches, y="team1", x="team2", hue="match_type", height=10, aspect=1.5)
```
