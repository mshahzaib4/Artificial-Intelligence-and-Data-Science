```python
import pandas as pd
import seaborn as sns
import pandas as np
import matplotlib.pyplot as plt
```

### Some Seaborn Matrix Plots

- **Heat map**
- **Cluster map**

```python
data = {
    'Shahzaib': {'Roll no': 10, 'Age': 20, 'class': 10},
    'Zohaib': {'Roll no': 11, 'Age': 10, 'class': 9},
    'Shahzaib Malik': {'Roll no': 22, 'Age': 21, 'class': 11}
}

data=np.DataFrame(data)
data
```

### `1.` **Creat Heat Map**

```python
plt.suptitle("Scatter Plot")
sns.heatmap(data, annot=True)
```

```python
sns.heatmap(data, annot=True, cmap='viridis')
```

```python
sns.heatmap(data, annot=True, cmap='viridis', linewidths=.5)
```
