```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
AirPassengers=pd.read_csv("AirPassengers time Series/AirPassengers.csv")
```

```python
AirPassengers.head()
```

```python
AirPassengers.shape
```

```python
AirPassengers.columns=["Date","Number of Passengers"]
```

```python
sns.set(rc={"figure.figsize":(16,6)})
sns.lineplot(data=AirPassengers,x="Date",y='Number of Passengers',color="Red")
plt.xticks([])
```

```python

decomposition = seasonal_decompose(AirPassengers['Number of Passengers'], model='multiplicative',period=12)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(16, 8))

plt.subplot(411)
plt.plot(AirPassengers['Number of Passengers'], label='Original',color="red")
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(trend, label='Trend',color="Green")
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(seasonal, label='Seasonal',color="Black")
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(residual, label='Residual',color="Brown")
plt.legend(loc='upper left')
plt.show()
```
