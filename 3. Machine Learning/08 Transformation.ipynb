```python
# import library
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
random_ = np.random.exponential(size=100,scale=2)
df_random_=pd.DataFrame(random_,columns=["Values"])
df_random_
```

```python
sns.histplot(data=df_random_['Values'],kde=True)
```

```python
from sklearn.preprocessing import power_transform , quantile_transform
```

```python

# Reshape for Box-Cox transformation
pt_boxcox= df_random_["Values"].values.reshape(-1, 1)
df_random_['values_boxcox'] = power_transform(pt_boxcox, method='box-cox', standardize=True)

# Reshape for Yeo-Johnson transformation
pt_yeo_johnsaon = df_random_["Values"].values.reshape(-1, 1)
df_random_['yeo_johnson']= power_transform(pt_yeo_johnsaon, method='yeo-johnson', standardize=True)

# Quantile_transf
df_random_["Quantile_transf"] = quantile_transform(df_random_[['Values']], n_quantiles=len(df_random_), random_state=0)
```

```python
df_random_.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10)) 

# Plot Box-Cox transformed data
sns.histplot(data=df_random_['values_boxcox'], kde=True, ax=axes[0,0])
axes[0,0].set_title('Box-Cox Transformation')

# Plot Yeo-Johnson transformed data
sns.histplot(data=df_random_['yeo_johnson'], kde=True, ax=axes[0,1])
axes[0,1].set_title('Yeo-Johnson Transformation')

sns.histplot(data=df_random_, kde=True ,ax=axes[1, 0])
axes[1, 0].set_title('First Hist Plot')


sns.histplot(data=df_random_['Quantile_transf'], kde=True ,ax=axes[1, 1])
axes[1, 1].set_title('Quantile_transft')
```
