 # **Cross Validation**

Cross-validation is a statistical method used in machine learning to assess the performance of a predictive model on an independent data sets.

Cross-validation involves splitting data into a training set and a validation set. The computer learns from the training set and is then tested on the validation set multiple times to ensure consistent performance.It's like giving it different quizzes to make sure it really understands the material.

**Types of CV**
- **k-fold cross-validation**
- **Train-test split**
- **Leave-one-out cross-validation (LOOCV)**
- **Stratified cross-validation**
- **Times series cross-validation**
- **Group cross-validation**

## **K-fold Cross Validation**
The model is trained on k-1 folds and evaluated on the remaining fold. This process is repeated k times, with each fold serving as the holdout or test set once.  The model is trained on k-1 folds and evaluated on the remaining fold. This process is repeated k times, with each fold serving as the holdout or test set once. 

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

scores = cross_val_score(GaussianNB(),iris.data,iris.target, cv=kfold, scoring='accuracy')

print("Average Accuracy:", scores)
print("Average Accuracy:", scores.mean())
print("Average Accuracy:", scores.std())
```

> **`range(1,7),range(1,8): `** If you use range(1, 7) and encounter an error, it's likely because the length of the scores list does not match the number of elements generated by the range.

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,6), scores, color='blue', linestyle='dashed', marker='o', markerfacecolor='red', markersize=12)
plt.title('K-Fold Cross-Validation Results')
plt.xlabel('Fold Number')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1.0])
plt.show()
```
