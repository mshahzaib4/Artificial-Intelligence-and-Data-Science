# Normalization
The process of normalization scales and standardizes data, usually to a range between 0 and 1 or with a mean of 0 and a standard deviation of 1.

- `L2 Normalization:`
Rescales each sample (row) to have unit norm. This type of normalization is often used when dealing with text data. The L2 norm is calculated as the square root of the sum of the squared vector values.

- `L1 Normalization:`
Also rescales each sample (row) but with a different approach, ensuring the sum of the absolute values is 1 in each row. The L1 norm is calculated as the sum of the absolute vector values.

```python
from sklearn.preprocessing import Normalizer
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]
print("Data",data)
print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
normal = Normalizer(norm='l1').fit_transform(data)
print("L1 Normalzation",normal)
print("\n~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~\n")
normal2 = Normalizer(norm='l2').fit_transform(data)
print("L1 Normalzation",normal2)
```

```python
# example dataset with skewed values
df_data = { "Values": [1,5,10,20,50,100,200,500,1000,2000,5000,10000,20000,50000,100000]}
df_data = pd.DataFrame(df_data)
df_data.head()
```

```python
log_data=np.log(df_data)
log_data.head()
```

```python
fig, axes = plt.subplots(1,2,figsize=(12, 5))
sns.histplot(data=df_data, kde=True,ax=axes[0])
axes[0].set_title("Data Plot Before Log")
sns.histplot(data=log_data, kde=True,ax=axes[1])
axes[1].set_title("Data Plot After Log")
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1, 2, figsize=(12,6))

sns.histplot(data=data,kde=True,ax=axes[0])
sns.histplot(data=normal,kde=True,ax=axes[1])

```
