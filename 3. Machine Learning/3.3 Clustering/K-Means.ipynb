# **k-Means Clustering:**
- k-Means Clustering is a type of unsupervised learning algorithm that groups similar data


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic=sns.load_dataset("iris")
```

```python
titanic.head()
```

```python
from sklearn.cluster import KMeans
```

```python
WCSS = []  

for i in range(1, 10):  
    Km = KMeans(n_clusters=i, random_state=42)  # Added random_state for reproducibility  
    Km.fit(titanic[["petal_length", "petal_width"]])  
    WCSS.append(Km.inertia_)  
```

```python
plt.scatter(titanic["petal_length"],titanic[ "petal_width"])
```

```python
plt.plot(range(1,10),WCSS)
```

```python
x = titanic[["petal_length", "petal_width"]].values  

KM = KMeans(n_clusters=4)   
y_means = KM.fit_predict(x)
```

```python
plt.scatter(x[y_means==0,0],x[y_means==0,1],color ="red")
plt.scatter(x[y_means==1,0],x[y_means==1,1],color ="green")
plt.scatter(x[y_means==2,0],x[y_means==2,1],color ="black")
```

# **Hierarchy Clustering**

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(18,6))
dend = shc.dendrogram(shc.linkage(titanic[["petal_length", "petal_width"]]))
```

# **Agglomerative Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
Cluster = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage='ward')
labels=Cluster.fit_predict(titanic[["petal_length", "petal_width"]])
```

```python
plt.scatter(titanic["petal_length"],titanic[ "petal_width"],c=Cluster.labels_,cmap="rainbow" )
```
