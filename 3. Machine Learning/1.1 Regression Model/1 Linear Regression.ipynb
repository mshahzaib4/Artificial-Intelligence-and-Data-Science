# Linear Regression
Linear Regression is a type of supervised machine learning algorithm that involves predicting continuous output with a consistent slope. It is commonly used for making predictions within a continuous range, such as sales or prices, rather than categorizing them into distinct groups like cat or dog.
## Types

`1.` Simple Linear Regression

`2.` Multiple Linear Regression

`3.` Polynomial Regression

`4.` Quantile Regression

`5.` Robust Regression and many more

**There are two main types**

- `Simple Linear Regression:` Uses one independent variable to predict a dependent variable.

- For Example : Y=MX+B 

- It models the relationship between one independent variable and the dependent variable with a straight line

- `Multiple Linear Regression:` Involves two or more independent variables predicting a dependent variable.

-Extending simple linear regression to multiple predictors involves modeling the relationship between multiple independent variables and the dependent variable.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Laod "Tips" dataset
plt.figure(figsize=(11,6))
Tips=sns.load_dataset("tips")
sns.scatterplot(data=Tips,x="total_bill",y="tip")
```

```python
X=Tips[["total_bill"]]
Y=Tips[["tip"]]
```

```python
# Split the data into traina and Test
X_train, X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
# call Model & train the model
model=LinearRegression().fit(X_train,Y_train)
```

```python
# predict
y_pred = model.predict(X_test)
y_pred
```

```python
# evaluate the model
print('MSE = ', mean_squared_error(Y_test, y_pred))
print('R2  = ', r2_score(Y_test, y_pred))
print('RMSE= ', np.sqrt(mean_squared_error(Y_test, y_pred)))
```

```python
# plot the model and data
plt.figure(figsize=(11,6))
plt.scatter(X_test, Y_test)
plt.plot(X_test, y_pred, color='red')
plt.show()
```

```python
# predicat Tip
total_bill = float(input("Enter something: "))
predicted_tip=model.predict([[total_bill]])

print(f"Predicted tip on a total bill of {total_bill} PKR is {predicted_tip}PKR\n")

```
