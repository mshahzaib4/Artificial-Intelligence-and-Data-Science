# Feature Encoding
`Feature encoding` is converting non-numeric data into a numerical format for machine learning algorithms.Feature encoding is particularly important when working with categorical data, which consists of discrete and often text-based values.
Feature encoding is like translating information for a computer

 There are several common techniques for feature encoding:

`1.` ` Label encoding` is a feature encoding technique that replaces categorical values with unique numerical labels, assigning each category a unique integer starting from 0 or 1.

For example, consider the colors "Red," "Green," and "Blue." In one-hot encoding : 
Red:   [1, 0, 0]
Green: [0, 1, 0]
Blue:  [0, 0, 1]

`2.` `One-hot encoding:` represents categorical variables as binary vectors, where each category is transformed into a binary vector of length N, with only one element as 1 and the rest as 0.

`3.` `Ordinal Encoding:` Assigning numerical values to categories based on their order or rank.

`4.` `Binary Encoding:` Encoding categorical features using binary code.

`5.` `Frequency encoding:` is a feature encoding technique where each category is replaced with its frequency in the dataset.

For example, consider a dataset with colors:

- Red: 5 occurrences
- Green: 8 occurrences
- Blue: 3 occurrences

In frequency encoding, the representation would be:

- Red: 5
- Green: 8
- Blue: 3

# Data Encoding Benefits

**`1.` Preservation of Information** Maintains essential details and relationships during encoding.

**`2.` Improved Model Performance** Well-encoded features enhance machine learning model performance.

 **`3.` Handling of Categorical Data** Enables the use of categorical variables in mathematice

 **`4.` Avoidance of Misinterpretation** Prevents misinterpretation of ordinal or nominal information for accurate representation.
 
 **`5.` Flexibility in Representation** Provides various encoding techniques to adapt to different data types and model requirements.

**`6.` Handling of Missing Values** Certain methods effectively handle missing values, improving data robustness.


```python
# import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

titanic_df=sns.load_dataset("titanic")
titanic_df.head()
```

```python
# Label encoder

from sklearn.preprocessing import LabelEncoder,OneHotEncoder, OrdinalEncoder
```

```python
titanic_df["Encoded_sex"]=LabelEncoder().fit_transform(titanic_df["sex"])
titanic_df.head()
```

```python
# One-Hot Encoding
titanic_df = sns.load_dataset("titanic")
one_hot_encoder = OneHotEncoder()
one_hot_encoded = one_hot_encoder.fit_transform(titanic_df[["who"]]).toarray()

# Creating new columns for each category
categories = one_hot_encoder.get_feature_names_out(["who"])
titanic_df[categories] = pd.DataFrame(one_hot_encoded, columns=categories)
titanic_df.head()
```

```python
# Ordinal Encoding
titanic_df=sns.load_dataset("titanic")
titanic_df["Ordinal_encoding"] = OrdinalEncoder(categories=[["First", "Second", "Third"]]).fit_transform(titanic_df[["class"]])
titanic_df.head()
```

```python
# Binary Encoding
from category_encoders import BinaryEncoder

# Assuming you have seaborn and pandas imported
titanic_df = sns.load_dataset("titanic")

# Applying Binary Encoder to the "who" column
BinaryEncoded_features = BinaryEncoder().fit_transform(titanic_df["who"])

# Adding the Binary Encoded features to the original DataFrame
titanic_df = pd.concat([titanic_df, BinaryEncoded_features])
titanic_df.head()
```
