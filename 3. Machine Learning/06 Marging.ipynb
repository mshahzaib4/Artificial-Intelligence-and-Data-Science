# `Merge, join, and concatenate`

The merge() function combines the data from two DataFrames by merging them together, using the specified method(s). You can use the parameters to determine which values to include in the merge.

- `merge():` The merge function is used to combine datasets based on a common column or index, or both.

- `concat():` The concat function is used to combine datasets either across rows or columns.

- `join():` The join function is used to combine datasets based on a key column or index.

```python
import pandas as pd
```

```python
# First data set
Data={'Name':["Shahzaib","Abduulah Gujjar","Abdullah","Zeshan Ali","Kiran","Rehman Ali"]
      ,'id':[1,2,3,4,5,6],'City':["Nwl",'Narowal','Lhr','Narowal','Nwl','Nwl']       
      }
Data=pd.DataFrame(Data)
Data.head()
```

```python
# Second data set
Data2={'Father':["Muhammad Yaqoob","Bashrat","Sttar","Gulam Hussain","Kiran","Rafaqat"]
      ,'id':[1,2,3,4,8,6],'City':["Nwl",'Narowal','Lhr','Narowal','Nwl','Nwl']       
      }
Data2=pd.DataFrame(Data2)
Data2.head()
```

```python
Merage=pd.merge(Data,Data2,on="id",how='inner')
Merage
```

# `Concatenate`

```python
concatenate=pd.concat([Data,Data2])
concatenate
```

```python
# Usse axis=1
concatenate=pd.concat([Data,Data2],axis=1)
concatenate
```
