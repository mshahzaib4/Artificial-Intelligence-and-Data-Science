```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Some Seaborn Relational Plots

- **Scatter plot**
- **Line plot**
- **Pair plot**

```python
data=pd.read_csv("Matches.csv")
data.head(2)
```

### `1.` **Creat Scatter Plot**

```python
plt.figure(figsize=(11, 5))
plt.suptitle("Scatter Plot")
sns.scatterplot(data=data,x="team1",y="winner_runs",hue="winner_runs")
```

### `2.` **Creat Line plot**

```python
plt.figure(figsize=(11, 5))
plt.suptitle("Line Plot")
sns.lineplot(data=data,x="team1",y="team2",hue="winner")
```
