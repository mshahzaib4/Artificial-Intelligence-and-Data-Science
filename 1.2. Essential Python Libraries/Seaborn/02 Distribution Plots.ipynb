### Some Seaborn Distribution Plots
- **Hist plot**
- **Kde plot**
- **Rug plot**
- **Ecdf plot**
- **Boxen plot**
- **Violin plot**
- **Qq plot**
- **Ridge plot**
- **Dis plot**

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data=pd.read_csv("deliveries.csv")
Data.head(2)
```

```python
Data.drop(['wides',	'noballs',	'byes',	'legbyes',	'penalty',	'wicket_type',	'player_dismissed',	'other_wicket_type','other_player_dismissed'],axis=1,inplace=True)
```

```python
Data.head(2)
```

# **`1.` Create Disp Plot**


```python
plt.figure(figsize=(15, 6))

plt.suptitle("Deliveries")
plt.xlabel("Match Season")
plt.ylabel("Match Venue")
sns.histplot(data=Data,x='match_id')
```

# **`2.` Create Kde Plot**


```python
plt.figure(figsize=(15, 6))

plt.suptitle("Deliveries")  # Set a title for the entire figure
plt.xlabel("Match Season")
plt.ylabel("Match Venue")

sns.kdeplot(data=Data, x='match_id', fill=True)  # Use kdeplot for KDE plot

plt.show()
```


# **`3.` Create Rug Plot**


```python

sns.displot(data=Data,x='match_id',kind='kde',rug=True)
```

# **`4.` Create Ecdf Plot**


```python
plt.figure(figsize=(15, 6))

plt.suptitle("Deliveries")
plt.xlabel("Match Season")
plt.ylabel("Match Venue")
sns.ecdfplot(data=Data, y="venue",  hue="innings")
plt.tight_layout()

```

# **`5.` Create Boxen Plot**


```python
plt.figure(figsize=(15, 10))

plt.suptitle("Deliveries")  # Set a title for the entire figure
plt.xlabel("Match Season")
plt.ylabel("Match Venue")
sns.boxenplot(data=Data, y="venue", x="season", hue="innings")
plt.tight_layout()

```

# **`6.` Create Violin Plot**


```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.violinplot(x='season', y='ball',hue="bowling_team", data=Data)
plt.title('Violin Plot')
```

## **`7.` Create Q-Q Plot (Quantile-Quantile Plot)**


```python
import statsmodels.api as sm
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sm.qqplot(x='season', y='ball',hue="bowling_team", data=Data)
plt.title('Violin Plot')
```

# **`2.` Create Residplot Plot**

- The **sns.residplot** function in Seaborn is specifically created to operate with numerical data.

```python

plt.figure(figsize=(15, 10))
plt.suptitle("Deliveries")
sns.residplot(data=Data, x='match_id', y='innings', lowess=True, scatter_kws={'s': 80}, line_kws={'color': 'red'})
```
