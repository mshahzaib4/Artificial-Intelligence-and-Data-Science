```python
import pandas as pd
import seaborn as sns
import pandas as np
import matplotlib.pyplot as plt
```

### Some Seaborn Matrix Plots

- **FacetGrid**
- **Cluster map**

```python
data=sns.load_dataset("tips")
data.head(2)
```

### `1.` **Scatterplot using FacetGrid**

```python
g = sns.FacetGrid(data,col="sex",hue="day")

# Map the heatmap
g.map(sns.scatterplot,"total_bill","size")
```

```python


g = sns.FacetGrid(data,col="day",hue="sex",col_wrap=2, height=4)

g.map(sns.scatterplot,"total_bill","size").add_legend()
```
